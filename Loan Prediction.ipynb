Filling the Nulls

```python
train_df.isnull().sum()
```

```python
train_df["Gender"].fillna(train_df["Gender"].mode()[0], inplace = True)
```

```python
gender_fill = test_df["Gender"].mode()[0]
test_df["Gender"].fillna(gender_fill, inplace = True)
```

```python
train_df.dropna(subset=["Married"], inplace = True)
```

```python
test_df.dropna(subset=["Married"], inplace = True)
```

```python
dependent_fill = test_df["Dependents"].mode()[0]
test_df["Dependents"].fillna(dependent_fill, inplace = True)
```

```python
dep_fill = train_df["Dependents"].mode()[0]
train_df["Dependents"].fillna(dep_fill, inplace = True)
```

```python
SE_fill = train_df["Self_Employed"].mode()[0]
train_df["Self_Employed"].fillna(SE_fill, inplace = True)
```

```python
SE_fill = test_df["Self_Employed"].mode()[0]
test_df["Self_Employed"].fillna(SE_fill, inplace = True)
```

```python
train_df["LoanAmount"].fillna(train_df["LoanAmount"].median(), inplace = True)
```

```python
test_df["LoanAmount"].fillna(test_df["LoanAmount"].median(), inplace = True)
```

```python
test_df["Loan_Amount_Term"].fillna(test_df["Loan_Amount_Term"].median(), inplace = True)
```

```python
train_df["Loan_Amount_Term"].fillna(train_df["Loan_Amount_Term"].median(), inplace = True)
```

```python
CH_fill = train_df["Credit_History"].mode()[0]
train_df["Credit_History"].fillna(CH_fill, inplace = True)
```

```python
CH_fill = test_df["Credit_History"].mode()[0]
test_df["Credit_History"].fillna(value=CH_fill, inplace = True)
```

# Building model by dividing the training data to train-test data

```python
from sklearn.model_selection import train_test_split
```

```python
train_set, test_set = train_test_split(train_df, test_size = 0.3, random_state = 35)
```

```python
train_x = train_set.drop(columns =["Loan_Status"], axis = 1)
train_y = train_set["Loan_Status"]
```

```python
test_x = test_set.drop(columns =["Loan_Status"], axis = 1)
test_y = test_set["Loan_Status"]
```

# Selection of model

```python
from catboost import CatBoostClassifier
```

```python
cate_features_index =np.where(train_x.dtypes != np.float)[0]
```

```python
model = CatBoostClassifier(iterations=50)
```

```python
model.fit(train_x, train_y, cat_features = cate_features_index, plot= False)
```

```python
pred1 = model.predict(test_x)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
matrix = confusion_matrix(test_y, pred1, labels = ["Y", "N"])
print("Confusion matrix: \n", matrix)
```

```python
matrix1 = classification_report(test_y, pred1, labels = ["Y", "N"])
print(matrix1)
```
